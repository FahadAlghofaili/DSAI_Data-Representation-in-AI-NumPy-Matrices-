```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
# Extract numerical features only
# Rows = Students (Samples)
# Columns = Features
X=(df[["raisedhands","VisITedResources","AnnouncementsView","Discussion"]])
```

```python
X = np.array(X)
print("Matrex",X)
print("Dimensions",X.ndim)
print("Sahpe",X.shape)
```

```python
#Scaler
print("Scaler",X[0,0])
```

```python
#Row vector
print("One row vector",X[0])
print("Dimensions",X[0].ndim)
print("Sahpe",X[0].shape)
```

```python
#Column vector
print("One column vector",X[:,0])
print("Dimensions",X[:,0].ndim)
print("Sahpe",X[:,0].shape)
```

```python
T = X.reshape(480, 4, 1)
print("Tensor shape",T.shape)
print("Tensor Dimensions",T.ndim)
```
